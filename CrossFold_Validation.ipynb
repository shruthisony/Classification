## Cross Fold Validation

### Step-1: Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection  import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import load_digits
```

### Step-2: Reading and understanding data

```python
digits = load_digits()
```

```python
print(digits.data)
print(digits.feature_names)
print(digits.target)
print(digits.target_names)
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

### Step-3: Model Builiding

```python
df_train, df_test = train_test_split(df, test_size=0.3, random_state=42)
```

```python
X_train = df_train.drop('target', axis=1)
y_train = df_train['target']

X_test = df_test.drop('target', axis=1)
y_test = df_test['target']
```

**Using Logistic Regression**

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

```python
print('Training Accuracy:', accuracy_score(y_train, y_train_pred))
print('Testing Accuracy:', accuracy_score(y_test, y_test_pred))
```

**Using Support Vector Machine**

**Using Random Forest Classifier**

### Step-4: Using Cross-fold Validation

```python
from sklearn.model_selection import cross_val_score
```

**Using Logistic Regression**

```python
scores_lr = cross_val_score(LogisticRegression(), X_train, y_train,cv=5)
print(scores_lr)
print('Mean:',np.mean(scores_lr))
```

**Using Support Vector Machine**

```python
scores_svm = cross_val_score(SVC(), X_train, y_train,cv=5)
print(scores_svm)
print('Mean:',np.mean(scores_svm))
```

**Using Random Forest Classifier**

```python
scores_rfc = cross_val_score(RandomForestClassifier(n_estimators=40), X_train, y_train,cv=5)
print(scores_rfc)
print('Mean:',np.mean(scores_rfc))
```

**Using Decision Tree Classifier**

```python
scores_dt = cross_val_score(DecisionTreeClassifier(), X_train, y_train,cv=5)
print(scores_dt)
print('Mean:',np.mean(scores_dt))
```

**So we can see that we can improve the accuracy scores by using cross-fold validation method**
