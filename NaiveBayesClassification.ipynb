### Step-1: Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

### Step-2: Reading and understanding data

```python
df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

### Step-3: Data Cleaning

```python
df.isnull().sum()/len(df)*100
```

```python
df = df.drop('Cabin',axis=1)
df.head()
```

```python
df.isnull().sum()/len(df)*100
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()/len(df)*100
```

### Step-4: Data Preparation

```python
df.head()
```

```python
df = df.drop(['PassengerId','Name','Ticket'],axis=1)
df.head()
```

```python
col1 = ['Sex','Embarked']
dummy1 = pd.get_dummies(df[col1],dtype = 'int64', drop_first=True)
dummy1.head()
```

```python
df = pd.concat([df,dummy1],axis=1)
df.drop(['Sex','Embarked'],axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
df_train, df_test = train_test_split(df,train_size=0.7,random_state=42)
```

```python
scalar_col = ['Pclass','Age','SibSp','Parch','Fare','Sex_male','Embarked_Q','Embarked_S']
```

```python
scalar = StandardScaler()
```

```python
df_train[scalar_col] = scalar.fit_transform(df_train[scalar_col])
df_train.head()
```

```python
df_test[scalar_col] = scalar.transform(df_test[scalar_col])
df_test.head()
```

### Step-5: Model Building and Predictions

```python
X_train = df_train.drop('Survived',axis=1)
y_train = df_train['Survived']

X_test = df_test.drop('Survived',axis=1)
y_test = df_test['Survived']
```

```python
nbc = GaussianNB()
nbc.fit(X_train,y_train)
```

```python
y_train_pred = nbc.predict(X_train)
y_test_pred = nbc.predict(X_test)
```

```python
print(accuracy_score(y_train,y_train_pred))
print(accuracy_score(y_test,y_test_pred))
```
